# それぞれの説明変数をもとに、弁当の販売数を予測するモデルを作成しよう

```python
df1.head()
```

データを見ると、弁当の売り上げに関係している要素は、弁当の種類、そのカロリー数、天気、気温と考えられます。したがって、これらの４つの変数を説明変数として、弁当の販売数を予測する回帰モデルを作成することを考えます。

データの中身を見るとわかるように、カロリーに関しては、NaN(欠損値)となっているものが含まれています。これが入ったままだと分析が非常に面倒です。

ですので、まずはこの欠損値を除去する必要があります。

欠損値を除去するためのDataFrameメソッドとして、dropnaと言ったものがあります。

こちらは、条件を指定した上で、その条件を満たすデータの欠損値を除去する役割を持っています。

今回、カロリーの欠損値を除去したいので、カラム名は「kcal」を指定してください。

```python
df1 = df1.dropna(subset=['kcal'])
```

```python
df1.head()
```

これで、カロリーが欠損しているデータが除外されました。これからの分析は全て、このデータを元に行なって行きます。

ところで、ここではカロリーの欠損値は除外されましたが、残念ながらまだ分析を行う上では不十分です。

なぜなら、天気や弁当の種類などのデータがまだ文字列になっているからです。こちらを数値に直さないと、機械学習の手法を適用することはできません。

ですのでまずは、この「天気」を数値ラベルに直すことを考えます。このデータで、天気は全部で7種類あり、順に"快晴","晴れ","薄曇","曇","雨","雷電","雪"が含まれています。

これらの７種類のデータ集を全て数値に直します。

こちらを文字列に変換するためには、データの前処理の章でも学んだstr.reshape()メソッドを使用します。

全ての天気情報を逐一変換することも可能ですが、手間がかかるのでプログラムで自動化する事を考えます。

次のコードを入力してください。

```python
weatherlist = df1["weather"].unique()
for i in range(len(weatherlist)):
    df1["weather"] = df1["weather"].replace(weatherlist[i], str(i))
```

```python
 df1["weather"]
```

```python
 df1["weather"].unique()
```

```python
len(weatherlist)
```

```python
df1
```

一見これで問題ないように思われるかもしれませんが、実はここで変換した数値は一見数値に見えますが、実は文字列のままです。ですので、次のコードを入力して浮動小数点数に直す必要があります。

```python
df1['weather'] = df1['weather'].astype(float)
```

```python
df1
```

```python
df1['name'].value_counts()
```

続いて、重複していない弁当の種類がどのくらい存在するか確認して行きます。次のコードを入力してください。



```python
df1['name'].unique()
```

```python
# 今回も天気と同様に、種類を数値ラベルに直して行きます。
# 次のコードを入力してください。

namelist = df1["name"].unique()
for i in range(len(namelist)):
    df1["name"] = df1["name"].replace(namelist[i], str(i))
```

```python
df1
```

```python
# 最後に、今回の学習で予測する変数である、「販売個数」が実はint型(整数型)
# であり、浮動小数点数表示になっておりません。ですので、
# こちらを適切なデータ型に直す必要があります。

df1['y'] = df1['y'].astype(float)
```

```python
df1
```

今回このデータを使用して具体的に行いたい分析は、

１.目的変数より、弁当の販売個数を予測する回帰モデルを作成する

2.弁当の販売個数を含めた全ての変数を目的変数とし、その弁当が売り切れたかどうかを判定する分類モデルを作成する

この２種類の手法を実行していくことが今回の分析の最終目標となります。

まずは、回帰モデルを作成することを試みて行きます。

まずは、今回のデータで使用する説明変数を取り出します。次のコードを入力してください。

```python
X = df1[['name','kcal', 'weather', 'temperature']]
```

```python
X.head()
```

```python
y = df1['y']
y.head()
```

```python
print(X.shape)
print(y.shape)
print(X)
print(y)
```

```python
# トレーニングデータとテストデータの分割の為の関数をインストール
from sklearn.cross_validation import train_test_split
# トレーニングデータとテストデータを70%:30%の割合で分割
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
# トレーニングデータのshapeを確認
print(X_train.shape)
print(y_train.shape)
# テストデータのshapeを確認
print(X_test.shape)
print(y_test.shape)
```

```python
# 線形回帰モデルの読み込み
from sklearn.linear_model import LinearRegression
```

```python
# 線形回帰モデルをインスタンス化
lr = LinearRegression()
```

```python
# トレーニングデータにもとづいて学習
lr.fit(X_train, y_train)
```

```python
# 係数
print(lr.coef_)
 # 切片 (誤差)
print(lr.intercept_)
```

```python
# 決定係数(トレーニングデータの場合)
print(lr.score(X_train, y_train))
```

```python
# 決定係数(テストデータの場合)
print(lr.score(X_test, y_test))
```

```python
# 必要なライブラリのインポート
from sklearn.cross_validation import train_test_split
import numpy as np
```

```python
X = df1[['y', 'name','kcal', 'weather', 'temperature']]
X.head()
```

```python
y = df1['soldout']
y.head()
```

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# ライブラリの読み込み
from sklearn.svm import SVC
```

```python
# SVMモデルのインスタンスを生成
svm = SVC(kernel='linear')
```

```python
# SVMのモデルにトレーニングデータセットを適合させる
svm.fit(X_train, y_train)
```

```python
# 正答率計算用のモジュールを読み込み
from sklearn.metrics import accuracy_score
# 予測結果を変数に代入
y_predicted = svm.predict(X_test)
# 正答率を算出
print('正答率', accuracy_score(y_test, y_predicted))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_predicted)
```

```python
# カーネルSVMを使用した場合の正解率
svm = SVC(kernel='rbf')
svm.fit(X_train, y_train)
y_predicted = svm.predict(X_test)
print('カーネルSVMの正答率', accuracy_score(y_test, y_predicted))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_predicted)
```
