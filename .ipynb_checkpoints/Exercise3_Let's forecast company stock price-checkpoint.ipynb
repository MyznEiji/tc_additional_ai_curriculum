データを見ると、株価に関しては様々な種類がありますが、今回は「終値」を予測するモデルを作成することを目標にします。

今回のデータで、説明変数は日付に相当するインデックス、目的変数は終値となります。

今までと同様に、説明変数、目的変数を取り出します。次のコードを入力してください。

```python

X = df['インデックス']
y = df['終値']
X
y
```

```python
# numpyの読み込み
import numpy as np

# matplotlibの読み込みを行う、pltという関数で呼び出せるようにする
import matplotlib.pyplot as plt
# matplotlibによるグラフ描画をJupyter Notebook上で実行できるようにする
%matplotlib inline
```

```python
plt.plot(X, y)
```

```python
from sklearn.preprocessing import PolynomialFeatures #多項式回帰のためのライブラリ
from sklearn.metrics import r2_score #決定係数のライブラリ
```

```python
X = df[['インデックス']].values #説明変数を指定
y = df['終値'].values #目的変数を指定
```

```python
# ライブラリより，線形回帰モデルを読み込む
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
#多項式回帰のための特徴量を作成する
quadrat = PolynomialFeatures(degree=2) #入力データを2次多項式に変換
cubic = PolynomialFeatures(degree=3) #入力データを3次多項式に変換
X_quad = quadrat.fit_transform(X) #学習のためにデータを２次元に整形
X_cubic = cubic.fit_transform(X) # 学習のためにデータを３次元に整形
```

```python
#線形回帰モデルを学習させる
lr = lr.fit(X, y)
y_lin_fit = lr.predict(X)# 学習済みモデルを用いて未知の値を予測する
linear_r2 = r2_score(y, lr.predict(X))
```

```python
#2次の多項式モデルの学習を行う
lr = lr.fit(X_quad, y) #二次多項式モデルと正解データの
y_quad_fit = lr.predict(quadrat.fit_transform(X))
quadratic_r2 = r2_score(y, lr.predict(X_quad))
```

```python
#3次の多項式モデルの学習を行う
lr = lr.fit(X_cubic, y)
y_cubic_fit = lr.predict(cubic.fit_transform(X))
cubic_r2 = r2_score(y, lr.predict(X_cubic))
```

最後に、線形回帰、二次、三次多項式回帰で学習したモデルの精度を比較します。

この章では、「r_2」という指標が出てきますが、こちらはカリキュラムでも学習した決定係数を意味しています。

カリキュラムの回帰と同様に、線形回帰、多項式回帰の双方で、決定係数の比較を行っていきます。

次のコードを入力することで、両者の比較が可能です。

まずは、元データをグラフ上にプロットします。

```python
# 元のグラフをプロットする
plt.scatter(X, y, label='training points', color='lightgray')
```

```python
plt.plot(X, y_lin_fit, 
         label='linear (d=1), $R^2=%.2f$' % linear_r2,# 線形回帰の場合の決定係数
         color='blue', 
         lw=2, 
         linestyle=':')
```

```python
plt.plot(X, y_quad_fit, 
         label='quadratic (d=2), $R^2=%.2f$' % quadratic_r2,# 二次多項式回帰の場合の決定係数
         color='red', 
         lw=2,
         linestyle='-')
```

```python
plt.plot(X, y_cubic_fit, 
         label='cubic (d=3), $R^2=%.2f$' % cubic_r2,#三次多項式回帰の場合の決定係数
         color='green', 
         lw=2, 
         linestyle='--')
```

```python
# 元のグラフをプロットする
plt.scatter(X, y, label='training points', color='lightgray')

plt.plot(X, y_lin_fit, 
         label='linear (d=1), $R^2=%.2f$' % linear_r2,# 線形回帰の場合の決定係数
         color='blue', 
         lw=2, 
         linestyle=':')


plt.plot(X, y_quad_fit, 
         label='quadratic (d=2), $R^2=%.2f$' % quadratic_r2,# 二次多項式回帰の場合の決定係数
         color='red', 
         lw=2,
         linestyle='-')


plt.plot(X, y_cubic_fit, 
         label='cubic (d=3), $R^2=%.2f$' % cubic_r2,#三次多項式回帰の場合の決定係数
         color='green', 
         lw=2, 
         linestyle='--')


plt.xlabel('% time index ')
plt.ylabel('final')
plt.legend(loc='upper right')
plt.show()
```
