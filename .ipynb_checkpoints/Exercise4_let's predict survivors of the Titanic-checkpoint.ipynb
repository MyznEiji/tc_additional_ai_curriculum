```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/we-b/datasets_for_ai/master/titanic.csv')
```

```python
df
```

問：読み込んだデータより、乗客の生存の有無を判定する分類モデルを作成してください。

ただし、今回適用する手法は、カリキュラムで学習した手法のうち、最低２つを選び、精度を比較するところまで終わらせてください。

具体的な手段としては、次のように進めましょう。

データの中身をご覧になっていただくと、一部欠損が含まれている箇所がございます。ですので、まずはこちらを除外していただく必要があります。欠損があるのは母親の年齢を表す「Age」, 乗船港を表す「Embarked」の二つです。ですのでまずはこれらのカラムの欠損値を除外してみましょう。

今回取り扱うデータには、一部データの形が数字ではない形式になってしまっているものがあります。ですので、まずはこちらを数字に直してください。

全てのデータを説明変数、目的変数に分けてください。

4.データをトレーニングデータ、テストデータに分けてください。

今回使用するモデルのライブラリを読み込みます。モデルはご自身で選択したものを読み込んでください。

5で読み込んだクラスをインスタンス化してください。

作成したモデルにトレーニングデータを読み込み、学習させてください。

テストデータに対する分類精度を求めてください。

```python
# まずは欠損値を除外します。
# 年齢を表す「Age」には欠損が含まれているので、こちらを取り除きます。
df = df.dropna(subset=['Age'])
```

```python
# 到着した港にも欠損値が存在します。次のコードを入力し、それを取り除きましょう。
df = df.dropna(subset=['Embarked'])
```

```python
df
```

```python
# 性別、港の種類は数値となっていないので、前処理を行うことで数値に変換します。
df['Sex'] = df['Sex'].replace('male','0').replace('female','1')
```

```python
df 
```

```python
# 見かけ上数値になっているように見えますが、
# まだ文字列の方のままです。ですので、浮動小数点数表示に変換します。


df['Sex'] = df['Sex'].astype(float)
```

```python
df
```

```python
df['Embarked'] = df['Embarked'].replace('C','0').replace('Q','1').replace('S','2')
```

```python
df
```

```python
df['Embarked'] = df['Embarked'].astype(float)
```

```python
# Pclass, SibSp, Parchを浮動小数点数表示に変換します。

df['Pclass'] = df['Pclass'].astype(float)
df['SibSp'] = df['SibSp'].astype(float)
df['Parch'] = df['Parch'].astype(float)
```

```python
df
```

```python
X = df[['Pclass','Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']]
X
```

```python
y = df['Survived']
y
```

```python
print(X.shape)
print(y.shape)
print(X)
print(y)
```

```python
# トレーニングデータとテストデータの分割の為の関数をインストール
from sklearn.cross_validation import train_test_split
# トレーニングデータとテストデータを70%:30%の割合で分割
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
# トレーニングデータのshapeを確認
print(X_train.shape)
print(y_train.shape)
# テストデータのshapeを確認
print(X_test.shape)
print(y_test.shape)
```

```python
# ライブラリの読み込み
from sklearn.linear_model import LogisticRegression
```

```python
# ロジスティック回帰のインスタンスを生成
lr = LogisticRegression()
# トレーニングデータをモデルに適合する
lr.fit(X_train, y_train)
```

```python
y_p = lr.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_p)
```

```python
print('誤分類の個数：',  (y_test != y_p).sum())
```

```python
# 特徴量毎の重みを出力
print(lr.coef_)
# 切片を出力
print(lr.intercept_)
```

```python
from sklearn.metrics import accuracy_score
print('正答率:', accuracy_score(y_test, y_p))
```

```python
# 必要なライブラリのインポート
from sklearn.cross_validation import train_test_split
import numpy as np
```

```python
X = df[['Pclass','Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']]
X
```

```python
y = df['Survived']
y
```

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# ライブラリの読み込み
from sklearn.svm import SVC
```

```python
# SVMモデルのインスタンスを生成
svm = SVC(kernel='linear')
```

```python
# SVMのモデルにトレーニングデータセットを適合させる
svm.fit(X_train, y_train)
```

```python
# 正答率計算用のモジュールを読み込み
from sklearn.metrics import accuracy_score
# 予測結果を変数に代入
y_predicted = svm.predict(X_test)
# 正答率を算出
print('正答率', accuracy_score(y_test, y_predicted))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_predicted)
```

```python
# カーネルSVMを使用した場合の正解率
svm = SVC(kernel='rbf')
svm.fit(X_train, y_train)
y_predicted = svm.predict(X_test)
print('カーネルSVMの正答率', accuracy_score(y_test, y_predicted))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_predicted)
```
