```python
import pandas as pd
from sklearn.model_selection import train_test_split
df = pd.read_csv("https://raw.githubusercontent.com/we-b/datasets_for_ai/master/birth_weight.csv")
```

```python
df
```

```python
X = df[['AGE','LWT', 'RACE', 'SMOKE', 'PTL', 'HT', 'UI']]
```

```python
X
```

```python
y = df['LOW']
```

```python
y
```

```python
print(X.shape)
print(y.shape)
print(X)
print(y)
```

```python
# トレーニングデータとテストデータの分割の為の関数をインストール
from sklearn.cross_validation import train_test_split
# トレーニングデータとテストデータを70%:30%の割合で分割
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
# トレーニングデータのshapeを確認
print(X_train.shape)
print(y_train.shape)
# テストデータのshapeを確認
print(X_test.shape)
print(y_test.shape)
```

```python
# 線形回帰モデルの読み込み
from sklearn.linear_model import LinearRegression
```

```python
# 線形回帰モデルをインスタンス化
lr = LinearRegression()
```

```python
# トレーニングデータにもとづいて学習
lr.fit(X_train, y_train)
```

```python
# 係数
print(lr.coef_)
 # 切片 (誤差)
print(lr.intercept_)
```

```python
# 決定係数(トレーニングデータの場合)
print(lr.score(X_train, y_train))
```

```python
# 決定係数(テストデータの場合)
print(lr.score(X_test, y_test))
```

```python
# 必要なライブラリのインポート
from sklearn.cross_validation import train_test_split
import numpy as np
```

```python
X = df[['AGE','LWT', 'RACE', 'SMOKE', 'PTL', 'HT', 'UI']]
X
```

```python
y = df['LOW']
y
```

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# ライブラリの読み込み
from sklearn.svm import SVC
```

```python
# SVMモデルのインスタンスを生成
svm = SVC(kernel='linear')
```

```python
# SVMのモデルにトレーニングデータセットを適合させる
svm.fit(X_train, y_train)
```

```python
# 正答率計算用のモジュールを読み込み
from sklearn.metrics import accuracy_score
# 予測結果を変数に代入
y_predicted = svm.predict(X_test)
# 正答率を算出
print('正答率', accuracy_score(y_test, y_predicted))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_predicted)
```

```python
# カーネルSVMを使用した場合の正解率
svm = SVC(kernel='rbf')
svm.fit(X_train, y_train)
y_predicted = svm.predict(X_test)
print('カーネルSVMの正答率', accuracy_score(y_test, y_predicted))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_predicted)
```
